```python
def preprocess_penguins(penguins_df):
    """
    Preprocess penguin dataset:
    - Combine date and time into a single datetime column.
    - Round datetime to daily granularity.
    - Handle numerical and categorical columns appropriately.
    """

    # Rename 'lat' and 'lon'
    penguins_df = penguins_df.rename(columns={'latitude_mean': 'latitude', 'longitude_mean': 'longitude'})

    penguins_df['date'] = pd.to_datetime(penguins_df['date_gmt']).dt.date

    # Drop useless columns
    penguins_df = penguins_df.drop(columns=['Unnamed: 0', 'date_gmt'])

    penguins_df['latitude'] = penguins_df['latitude'].round(2)
    penguins_df['longitude'] = penguins_df['longitude'].round(2)

    # # ***ADD PHILIPP'S DATA ENGINEEERING***

    # # Combine date and time into a single datetime column
    # penguins_df['datetime'] = pd.to_datetime(penguins_df['date_gmt'] + ' ' + penguins_df['time_gmt'])
    
    # # Extract the date only
    # penguins_df['date'] = penguins_df['datetime'].dt.date
    
    # # Define numerical and categorical columns
    # # remove ID columns from numerical columns and put in categorical
    # numerical_columns = penguins_df.select_dtypes(include=['number']).columns.tolist()
    # numerical_columns = list(set(numerical_columns))  # Remove duplicates if latitude/longitude overlap
    # print(numerical_columns)
    # categorical_columns = list(set(penguins_df.columns) - set(numerical_columns) - {'datetime', 'date_gmt', 'time_gmt', 'date'})
    # print(categorical_columns)

    # # Group by date and location
    # aggregation = {col: 'mean' for col in numerical_columns}  # Use mean for numerical columns
    # aggregation.update({col: lambda x: x.mode().iloc[0] if not x.mode().empty else None for col in categorical_columns})  # Use mode for categorical columns
    # penguins_grouped = penguins_df.groupby(['date']).agg(aggregation).reset_index()

    # # Round numerical columns to 2 decimal places
    # for col in numerical_columns:
    #     if col in penguins_grouped.columns:  # Check if the column exists
    #         penguins_grouped[col] = penguins_grouped[col].round(2)
    
    return penguins_df



def preprocess_era5(era5_df):
    """
    Preprocess ERA5 weather dataset:
    - Align time format to date only.
    - Average data by date and coordinates if necessary.
    """
    # Convert time to date format
    era5_df['date'] = pd.to_datetime(era5_df['time']).dt.date

    era5_df['latitude'] = era5_df['latitude'].round(2)
    era5_df['longitude'] = era5_df['longitude'].round(2)
    
    # # Group by date and coordinates, calculate mean for numerical columns
    # agg_columns = ['latitude', 'longitude']
    # era5_grouped = era5_df.groupby(['date'] + agg_columns).mean().reset_index()
    
    return era5_df

def preprocess_sea_ice(sea_ice_df):
    """
    Preprocess sea ice dataset:
    - Expand monthly sea ice data to daily granularity.
    """
    sea_ice_df['date'] = pd.to_datetime(sea_ice_df['time']).dt.date
    sea_ice_df = sea_ice_df.drop(columns=['time'])

    # Rename 'lat' and 'lon' columns in sea_ice_df to 'latitude' and 'longitude'
    sea_ice_df = sea_ice_df.rename(columns={'lat': 'latitude', 'lon': 'longitude'})

    sea_ice_df['latitude'] = sea_ice_df['latitude'].round(2)
    sea_ice_df['longitude'] = sea_ice_df['longitude'].round(2)

    # # Generate a complete daily date range between the first and last date in the dataset
    # full_date_range = pd.date_range(start=sea_ice_df['time'].min(), 
    #                                 end=sea_ice_df['time'].max(), 
    #                                 freq='D')

    #  # Generate all combinations of coordinates and the full date range
    # coords = sea_ice_df[['lat', 'lon']].drop_duplicates()
    # expanded_grid = pd.MultiIndex.from_product(
    #     [coords['lat'], coords['lon'], full_date_range],
    #     names=['lat', 'lon', 'time']
    # ).to_frame(index=False)

    # # Merge the expanded grid with the original data
    # sea_ice_df = pd.merge(expanded_grid, sea_ice_df, on=['lat', 'lon', 'time'], how='left')

    # # Interpolate numeric columns linearly, forward and backward fill non-numeric data
    # for col in sea_ice_df.columns:
    #     if sea_ice_df[col].dtype in ['float64', 'int64']:
    #         sea_ice_df[col] = sea_ice_df[col].interpolate(method='linear')
    #     else:
    #         sea_ice_df[col] = sea_ice_df[col].fillna(method='ffill').fillna(method='bfill')

    return sea_ice_df


def preprocess_and_merge(penguins_df, era5_df, sea_ice_df):
    """
    Preprocess datasets and merge them by date and coordinates, applying gradient interpolation for null ice_conc values.
    """

    print("Merging datatets")
    # Merge datasets on date, latitude, and longitude
    merged_df = penguins_df.merge(
        era5_df, on=['date', 'latitude', 'longitude'], how='left'
    ).merge(
        sea_ice_df, on=['date', 'latitude', 'longitude'], how='left'
    )

    print("Sorting values")
    # Sort values for consistency
    merged_df = merged_df.sort_values(by=['latitude', 'longitude', 'date', 'track_id'])

    print("Filling ice_conc missing values")
    # Fill missing `ice_conc` values with gradient interpolation
    merged_df['ice_conc'] = merged_df.groupby(['latitude', 'longitude'])['ice_conc'].transform(
        lambda group: group.interpolate(method='linear')
    )

    merged_df = merged_df.reset_index()

    return merged_df


def main(penguins_df, era5_df, sea_ice_df):
    """
    Main function to preprocess and merge all datasets.
    """
    # Preprocess each dataset
    print("Pre processing penguins data")
    penguins_cleaned = preprocess_penguins(penguins_df)
    print("Pre processing era5 data")
    era5_cleaned = preprocess_era5(era5_df)
    print("Pre processing sea ice data")
    sea_ice_cleaned = preprocess_sea_ice(sea_ice_df)
    
    # Merge datasets
    final_dataset = preprocess_and_merge(penguins_cleaned, era5_cleaned, sea_ice_cleaned)
    
    return final_dataset
```
