## Penguin's Data

```python
directory = Path("..\data\datasets_penguins")
target_extension = ".csv"  # Change this to the desired file format

all_dataframes = []

# Iterate through all files in the directory
for file in directory.iterdir():
    if file.is_file() and file.suffix == target_extension:  # Check if it's a file
        print(f"File: {file.name}")
        df = pd.read_csv(f"..\data\datasets_penguins\{file.name}")
        all_dataframes.append(df)  # Add the DataFrame to the list

# Concatenate all DataFrames into one
combined_dataframe = pd.concat(all_dataframes, ignore_index=True)

combined_dataframe
```
